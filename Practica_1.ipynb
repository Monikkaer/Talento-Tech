```python
# Ejercicio 1: Mostrar las primeras filas del Dataframe
print('Ejercicio 1: Mostrar las primeras filas del Dataframe')
print(df.head(10))
```

```python
# Ejercicio 2:Calcular el total de ventas por región
print('Ejercicio 2:Calcular el total de ventas por región')
ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
print('\n')
```

```python
# Ejercicio 3: Filtrar el Dataframe para mostrar solo las ventas realizadas por David
print('Filtrar el Dataframe para mostrar solo las ventas realizadas por David')
ventas_david = df[df['Vendedor'] == 'David']
print(ventas_david)
print('\n')
```

```python
# Ejercicio 4: Calcular el promedio de ventas por producto
print('Calcular el promedio de ventas por producto')
ventas_por_producto = df.groupby('Producto')['Ventas'].mean()
print(ventas_por_producto)
print('\n')
```

```python
#Ejercicio 5: Ordenas el Dataframe por ventas en orden descendente
print('Ordenas el Dataframe por ventas en orden descendente')
df_ordenado = df.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
```

```python
# Ejercicio 6: Filtrar las filas con ventas mayores a 10000
print ('Filtrar las filas con ventas mayores a 10000')
df_filtrado = df[df['Ventas']>10000]
print(df_filtrado)
print('\n')
```

```python
#Ejercicio 7: Ordenas el Dataframe por ventas en orden descendente
print('Ordenas el Dataframe por ventas en orden descendente')
df_ordenado = df.sort_values(by='Ventas', ascending=True)
print(df_ordenado)
```

```python
#Ejercicio 8: Eliminar la columna "Vendedor"
print('Eliminar la columna "Vendedor"')
df_sin_vendedor = df.drop(columns=['Vendedor'])
print(df_sin_vendedor)
print('\n')
```

```python
# Ejercicio 9: Calcular el total de ventas por región
print('Calcular el total de ventas por región')
df_ventas_region = df.groupby('Region')['Ventas'].sum()
print(df_ventas_region)
print('\n')
```

```python
# Ejercicio 10: Calcular el total de las ventas por producto
print('Calcular el total de las ventas por producto')
df_ventas_por_producto = df.groupby('Producto')['Ventas'].sum()
print(df_ventas_por_producto)
print('\n')
```
